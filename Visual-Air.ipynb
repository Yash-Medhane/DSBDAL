```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/airquality.csv")
data.head(5)
```

```python
station_aqi = data.groupby('AQI_Bucket')['AQI'].mean()
plt.plot(station_aqi.index,station_aqi.values)
plt.show()
```

```python
plt.figure(figsize=(6,10))
plt.hist(data['Date'],bins=100)
plt.show()
```

```python
station_aqi = data.groupby('AQI_Bucket')['AQI'].mean().sort_values()
plt.bar(station_aqi.index,station_aqi.values)
plt.show()
```

```python
plt.figure(figsize=(6,10))
data.boxplot(column='PM2.5',by='AQI_Bucket')
plt.ylabel('PM2.5')
plt.xlabel('AQI_Bucket')
plt.show()
```

```python
df = data.drop(['StationId','Date','AQI_Bucket'],axis=1)
```

```python
df.head(5)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),annot=True)
plt.show()
```
