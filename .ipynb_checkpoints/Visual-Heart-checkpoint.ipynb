```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('heartdisease.csv')
data.head(5)
```

```python
chol_by_age = data.groupby('ChestPain')['Chol'].mean()
plt.plot(chol_by_age.index,chol_by_age.values)
plt.show()
```

```python
plt.hist(data['Age'],bins=60,color='green')
plt.grid('show')
```

```python
thal_count = data['Thal'].value_counts()
plt.bar(thal_count.index,thal_count.values)
plt.show()
```

```python
plt.figure(figsize=(5,50))
data.boxplot(column='Chol', by='Sex')
plt.show()
```

```python
plt.hist(data['Sex'],bins = 10,color='yellow')
plt.show()
```

```python
data.head(5)
```

```python
df = data.drop(['ChestPain','Thal','AHD'],axis=1)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(), annot=True)
```

```python
sns.distplot(data['RestBP'])
plt.show()
```

```python
sex_count = data['Sex'].value_counts()
sns.barplot(sex_count)
plt.show()
```

```python
sns.scatterplot(x='ChestPain',y='Chol',data = data)
plt.show()
```
